## Step 4: Vet Data Sources

Use the cells below to load the data, inspect it, and perform some exploratory data analysis.  

Add more cells (`Insert > Insert Cell Below`) if you want additional cells.

```python
import pandas as pd
import numpy as np
from sklearn import tree

with open('heart_failure_clinical_records_dataset.csv') as file:
    df = pd.read_csv(file)
```

```python
df.head()
```

```python
df.corr()
```

```python
# I will use a decision tree to see how all the variables (besides time) affect death
# The first step is to split the data into training and test sets
train = df.sample(frac=8/10, random_state = 1)
test = df.drop(train.index)

train_X = np.asarray(train[['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'ejection_fraction',
                            'high_blood_pressure','platelets','serum_creatinine', 'serum_sodium', 'sex', 'smoking']])
train_y = np.asarray(train.DEATH_EVENT)
test_X = np.asarray(test[['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'ejection_fraction',
                            'high_blood_pressure','platelets','serum_creatinine', 'serum_sodium', 'sex', 'smoking']])
test_y = np.asarray(test.DEATH_EVENT)
```

```python
# Fit the model
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(train_X, train_y )
```

```python
# Predict death using model
y_pred = clf.fit(train_X, train_y).predict(train_X)
print("Number of mislabeled points out of a total {} points : {}"
      .format(len(train),(train_y != y_pred).sum()))
```

```python
y_pred = clf.fit(train_X, train_y).predict(test_X)
print("Number of mislabeled points out of a total {} points : {}"
      .format(len(test),(test_y != y_pred).sum()))
```

```python
col_names = ['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'ejection_fraction',
             'high_blood_pressure','platelets','serum_creatinine', 'serum_sodium', 'sex', 'smoking']
z = zip(col_names, clf.feature_importances_)
list(z)
```

```python
import matplotlib.pyplot as plt
from plotnine import *
import seaborn as sns
```

```python
# First, I will make a pair plot of the remaining variables
# This is similar to a correlation matrix, but with scatterplots instead of correlations
sns.pairplot(df2[["DEATH_EVENT", "ejection_fraction", "platelets", "serum_creatinine", "serum_sodium"]])
```
