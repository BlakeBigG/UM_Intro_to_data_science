**Exercise 1**: *Read in the `congress-terms.csv` file and name it `df`.*

```r
# Exercise 1 code goes here
# -------------------------
df <- read.csv('/dsa/data/all_datasets/congress-terms.csv')
head(df)
```

**Exercise 2**: On average, does the house or the senate have older members? What statistics could you use to find this?

```r
# Exercise 2 code goes here
# -------------------------
house <- subset(df, chamber == 'house')
mean(house$age)
median(house$age)

senate <- subset(df, chamber == 'senate')
mean(senate$age)
median(senate$age)
```

**Exercise 3**: Who is the oldest person in this dataset? The youngest?

```r
# Exercise 3 code goes here
# -------------------------
df_sorted <- df[order(df$age),]
head(df_sorted, 1)
tail(df_sorted, 1)

```

**Exercise 4**: Who is the oldest and youngest person per political party?

```r
# Exercise 4 code goes here
# -------------------------
library(dplyr)

df %>%
    group_by(party) %>%
    filter(age == min(age))

df %>%
    group_by(party) %>%
    filter(age == max(age))
```

There is a column in this data frame called `bioguide`. This is a unique id for each individual in the data set. As you are probably aware, many politicians have multiple terms.

**Exercise 5**: Who is the politician with the most terms?

```r
# Exercise 5 code goes here
# -------------------------
#This step finds the ID code for the polititian that has the most terms
frequency_df <- as.data.frame(table(df$bioguide))
frequency_df_sorted <- frequency_df[order(frequency_df$Freq),]
tail(frequency_df_sorted, 1)
```

```r
df[df$bioguide == "D000355",]
```

**Exercise 6**: Who has had the most Senate terms? House terms?

```r
# Exercise 6 code goes here
# -------------------------
house <- subset(df, chamber == 'house')
frequency_house <- as.data.frame(table(house$bioguide))
frequency_house_sorted <- frequency_house[order(frequency_house$Freq),]
tail(frequency_house_sorted, 1)

senate <- subset(df, chamber == 'senate')
frequency_senate <- as.data.frame(table(senate$bioguide))
frequency_senate_sorted <- frequency_senate[order(frequency_senate$Freq),]
tail(frequency_senate_sorted, 1)
```

**Challenge Exercise 1**: Using the mean, which party on average holds a higher number of terms per politician? 

```r
# Challenge Exercise 1 code goes here
# -----------------------------------
republican <- subset(df, party == 'R')
mean(table(republican$bioguide))

democrat <- subset(df, party == 'D')
mean(table(democrat$bioguide))
```

**Challenge Exercise 2**: Using the median age, what condition has the strongest linear relationship: Congress Number and Republican House age, Congress Number and Republican Senate age, Congress Number and Democratic House age, or Congress Number and Democratic Senate age.

```r
# Challenge Exercise 2 code goes here
# -----------------------------------
house_republican <- subset(republican, chamber == 'house')
house_republican <- house_republican %>%
    group_by(congress) %>%
    summarize(age = median(age))
cor(house_republican$congress, house_republican$age)

house_democrat <- subset(democrat, chamber == 'house')
house_democrat <- house_democrat %>%
    group_by(congress) %>%
    summarize(age = median(age))
cor(house_democrat$congress, house_democrat$age)

senate_republican <- subset(republican, chamber == 'senate')
senate_republican <- senate_republican %>%
    group_by(congress) %>%
    summarize(age = median(age))
cor(senate_republican$congress, senate_republican$age)

senate_democrat <- subset(democrat, chamber == 'senate')
senate_democrat <- senate_democrat %>%
    group_by(congress) %>%
    summarize(age = median(age))
cor(senate_democrat$congress, senate_democrat$age)
```

**Exercise 7**: Plot the relationship with the strongest linear relationship to visualize the results.

```r
# Exercise 7 code goes here
# -------------------------
plot(house_democrat$congress, house_democrat$age)
```
